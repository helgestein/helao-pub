```python
import math
```

```python
def convert_current_to_range(current):
    sign = -1 if current < 0 else 1
    magnitude = math.floor(math.log10(abs(current)))
    normalized_value = abs(current) * 10 ** -(magnitude)
    if normalized_value > 5:
        magnitude += 1
        normalized_value *= 0.1
    return sign * normalized_value, magnitude
```

```python
current = -2.9673e-6  # Should output -0.6 and -2 (or equivalently 10 mA)
normalized_value, magnitude = convert_current_to_range(current)
print(f"Normalized Value: {normalized_value}, Range: 10^{magnitude} A")
```
